```python
data_1=pd.DataFrame(np.arange(12).reshape((4,3)),
                   index=[['a', 'a', 'b', 'b'], [1, 2, 1, 2]],
                   columns=[['Ohio', 'Ohio', 'Colorado'],
                            ['Green', 'Red', 'Green']])
data_1
```

```python
data_1.index.names=['key1','key2'] 
#data_1.index.names=['state','color'] 
data_1
```

```python
data_1.index.names=['state','color'] 
data_1
```

```python
data_1['Colorado']
```

```python
data_1['Ohio']
```

```python
#data_1.swaplevel('key1','key2')
```

```python
data_1.swaplevel('state','color')
```

```python
data_1.sort_index(level=1)
```

```python
data_1.swaplevel(0,1).sort_index(level=0)
```
