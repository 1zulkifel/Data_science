## reguler expression and vectorized function

```python
import re
import pandas as pd
import numpy as np
data={'pop':'poohpool@signet.com','mei': 'meixin_kok@hotmail.com', 
      'nagai':'ngai_nicole@hotmail.com','isabelle': 'isabelle_gal@hotmail.com' ,
      'isn':'insyirah_powergals@hotmail.com',}
#data=[ 'little_princess-angelhotmail.com;',
 #     'ifah-duff@hotmail.com;',' tweety_butt@hotmail.com;', 
  #    'choco_ela@hotmail.com;',' princessdyanah@hotmail.com;',' isabelle_gal@hotmail.com',]
data=pd.Series(data)
data
pattern='([A-Z0-9._\%+-]+)@([A-Z0-9.-])\\.([A-z]{2-4})'
data.str.findall(pattern,flags=re.IGNORECASE)
matches=data[data.str.match(pattern,flags=re.IGNORECASE)]
print(matches)
```

# Hierarchical indexing

```python
import pandas as pd 
import numpy as np
hdata=pd.Series(np.random.randn((9)),
               index=[['a','a','a','b','b','c','c','c','d'],
                     [1,2,3,2,5,2,7,8,9]])
hdata
```

```python
hdata.loc['b']
```

```python
hdata.loc[:,2]
```

```python
hdata.loc[['b','d']]
```

```python
hdf=hdata.unstack()
hdf
#clean=hdf.replace(np.nan,0.11)
#clean
```

```python
hdata.unstack().stack()#for orignal position of data
```

```python
#hirarchical data frame
dfh=pd.DataFrame(np.arange(12).reshape(4,3),
                index=[['a','a','b','b'],[1,2,1,2]],
                columns=[['me','you','me'],['green','red','green']])
dfh.index.names=['key1','key2']
dfh.columns.names=['state','color']
dfh
```

```python
dg=dfh.unstack()
dg
```

```python
dfhswp=dfh.swaplevel('key1','key2')
dfhswp
```

```python
dfhswp.unstack()
```

```python
#dfh['me']
dfh['you']
```

```python
dfh.sort_index(level=0)
```

```python
dfh.swaplevel(0,1).sort_index(level=1)
```
