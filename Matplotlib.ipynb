```python
import numpy as np
import matplotlib.pyplot as plt
ages_x =[20,21,22,23,24,25,26,27]
sal_y = [112,225,350,388,533,656,700,872]
x_index=np.arange(len(ages_x ))
width=0.30

plt.style.use('seaborn-darkgrid')
plt.bar(x_index-width,sal_y,color='b',width=width,linewidth=3,linestyle='-.',label='sal_y')

#write code below
sal_y1 = [212,312,455,566,675,786,886,867]
plt.bar(x_index,sal_y1,color='r',width=width,linewidth=4,linestyle='-.',label='sal_y1' )

#write code 
sal_y2=[331,318,350,450,399,390,441,452]
plt.bar(x_index+width,sal_y2,color='y',width=width,linewidth=3,linestyle='-',label='sal_y2')

plt.xlabel('ages')
plt.ylabel('salary in $')
plt.title('Age vs Salary bar graph')
plt.legend()
plt.xticks(ticks=x_index,labels=ages_x)
plt.grid()
plt.tight_layout()
 
plt.show()
```
